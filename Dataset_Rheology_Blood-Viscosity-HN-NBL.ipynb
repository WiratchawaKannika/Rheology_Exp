```python
def checkFolderClasses(df_SSD_Backup):
    folderOut = []
    df_SSD_Backup_ = pd.DataFrame(columns=list(df_SSD_Backup.columns))
    for cls in classes:
        print(f"[INFO]: For {cls} Classes")
        cls_df = df_SSD_Backup[df_SSD_Backup["classes"]==cls]
        set_folder = list(set(cls_df["image_folder"]))
        print(f"HAvE {len(set_folder)} Folder")
        for folder in set_folder:
            folder_i = cls_df[cls_df["image_folder"]==folder]
            #print(folder_i.shape[0])
            if folder_i.shape[0] < 2001 :
                #print(folder_i)
                folderOut.append(folder)
                folder_i["complete"] = "no"
                df_SSD_Backup_ = df_SSD_Backup_.append(folder_i)
            else:
                folder_i["complete"] = "yes"
                df_SSD_Backup_ = df_SSD_Backup_.append(folder_i)
        print("="*125)
    
    return df_SSD_Backup_, folderOut
```

```python
def create_CodeSubclass(dataset_SSD):

    '''
    subclass
    Patients_‡∏ï‡∏±‡∏î‡∏°‡πâ‡∏≤‡∏°: Splenectomy
    Patients_‡πÑ‡∏°‡πà‡∏ï‡∏±‡∏î‡∏°‡πâ‡∏≤‡∏°: No_Splenectomy
    Normal
    '''

    df_dictCode = pd.read_csv("/home/kannika/codes_AI/Rheology_Blood/All_lab_results_Blood_viscosity_work3-subclass.csv")
    lst_Code = df_dictCode["Code"].tolist()
    lst_sheet = df_dictCode["sheet"].tolist()
    dict_code = zip(lst_Code, lst_sheet)
    output_dict = dict(dict_code)
    ## 1. map Code
    code_ = []
    for i in range(len(dataset_SSD)):
        image_folder = dataset_SSD['image_folder'][i]
        code = image_folder.split("_")[0]
        code_.append(code)
    dataset_SSD["Code"] = code_
    ## 2. map Splenectomy
    dataset_SSD["subclass"]= dataset_SSD['Code'].map(output_dict)
    print(dataset_SSD.shape)

    return dataset_SSD
```
