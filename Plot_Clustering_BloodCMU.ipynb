```python
def plot4group(result_4cluster_, columnName):

    set_sheet = list(set(result_4cluster_[columnName]))
    print(len(set_sheet))
    set_sheet
    #lt.figure(figsize=(12, 8))
    fig = plt.figure(figsize=(12, 8))
    handles_markers = []
    markers_labels = []
    colors_labels = []
    for s in set_sheet:
        result0 = result_4cluster_[result_4cluster_[columnName]==s]
        if s == 0.0:
            colors="red"
            markers="^"
        elif s == 1.0:
            colors="green"
            markers="8"
        elif s == 2.0:
            colors="blue"
            markers="s"
        elif s == 3.0:
            colors="fuchsia"
            markers=">"
        markers_labels.append(markers)
        colors_labels.append(colors)
        print(f"[INFO] Cluster : {s} with Shape ==> {result0.shape[0]}")
        set_code = list(set(result0["Code"]))
        for sc in set_code:
            df_SetCode = result0[result0["Code"]==sc].reset_index(drop=True)    
            x = df_SetCode["x"].tolist()
            y = df_SetCode["y"].tolist()

            #plt.style.use('seaborn')
            plot_handles = plt.plot(x, y, ls=":", marker=markers, markersize=5,lw=2, markerfacecolor="None",color=colors)
            markers_labels.append(s)
            handles_markers.append(handles_markers)
           #plt.plot(x, y, ls=":", marker="8", markersize=4, lw=2, markerfacecolor="None",color=colors)
           #plt.plot(t_x, list_preEff, ls=":", marker="^", markersize=4, lw=2, markerfacecolor="None", label="EffNetB7 Model")
    #         plt.plot(t_x, list_meanM, ls=":", marker="s", markersize=4, lw=2, markerfacecolor="None", label="MSDT Algorithm", color="red")

    plt.xlabel("x", fontsize=16)
    plt.ylabel("y", fontsize=16)
    plt.title('The Plot Of The Clusters: k = 4', fontsize=18)

    import matplotlib.lines as mlines
    red_line = mlines.Line2D([], [], color='red',markersize=5, label='Group 1', linestyle=":", marker="^", markerfacecolor="None")
    green_line = mlines.Line2D([], [], color='green', markersize=5, label='Group 2', linestyle=":", marker="8", markerfacecolor="None")
    blue_line = mlines.Line2D([], [], color='blue', markersize=5, label='Group 3', linestyle=":", marker="s", markerfacecolor="None")
    gold_line = mlines.Line2D([], [], color='fuchsia', markersize=5, label='Group 4', linestyle=":", marker=">", markerfacecolor="None")

    handles = [red_line,green_line, blue_line, gold_line]
    labels = [h.get_label() for h in handles] 

    fig.legend(handles=handles, labels=labels, borderaxespad=5)

    plt.show()
    
    return 
```

```python
import matplotlib.pyplot as plt

## Sheet 
set_sheet = list(set(pcaK3['sheet']))
set_cluster = list(set(pcaK3['Clusterk3']))
print(len(set_sheet))
set_sheet
#lt.figure(figsize=(12, 8))
fig = plt.figure(figsize=(12, 8))
handles_markers = []
markers_labels = []
colors_labels = []
for s in set_sheet:
    result0 = pcaK3[pcaK3["sheet"]==s]
    if s == "Patients_NoCutSpleen":
        colors="red"
        markers="8"
    elif s == "Patients_CutSpleen":
        colors="green"
        markers="8"
    elif s == "Normal":
        colors="blue"
        markers="8"
    markers_labels.append(markers)
    colors_labels.append(colors)
    print(f"[INFO] Sheet : {s} with Shape ==> {int(result0.shape[0]/3)} Cases")
    set_code = list(set(result0["Code"]))
    for sc in set_code:
        df_SetCode = result0[result0["Code"]==sc].reset_index(drop=True)    
        x = df_SetCode["x"].tolist()
        y = df_SetCode["y"].tolist()
    
        #plt.style.use('seaborn')
        plot_handles = plt.plot(x, y, ls=":", marker=markers, markersize=15,lw=2, markerfacecolor="None",color=colors)
        markers_labels.append(s)
        handles_markers.append(handles_markers)
print("*"*125)
### Cluster
for c in set_cluster:
    result00 = pcaK3[pcaK3["Clusterk3"]==c]
    if c == 0.0:
        colors_k="red"
        markers_k="s"
    elif c == 1.0:
        colors_k="green"
        markers_k="s"
    elif c == 2.0:
        colors_k="blue"
        markers_k="s"
    print(f"[INFO] Sheet : Cluster Group {int(c+1)} with Shape ==> {int(result00.shape[0]/3)} Cases")
    set_code = list(set(result00["Code"]))
    for sc in set_code:
        df_SetCode = result00[result00["Code"]==sc].reset_index(drop=True)    
        x_k = df_SetCode["x"].tolist()
        y_k = df_SetCode["y"].tolist()
    
        #plt.style.use('seaborn')
        plot_handles = plt.plot(x_k, y_k, ls=":", marker=markers_k, markersize=5,lw=2, markerfacecolor="None",color=colors_k)
       
        
plt.xlabel("x", fontsize=16)
plt.ylabel("y", fontsize=16)
plt.title('The Plot Of The Clusters', fontsize=18)

import matplotlib.lines as mlines
red_line = mlines.Line2D([], [], color='red',markersize=5, label='Patients No Cut Spleen', linestyle=":", marker="8", markerfacecolor="None")
green_line = mlines.Line2D([], [], color='green', markersize=5, label='Patients Cut Spleen', linestyle=":", marker="8", markerfacecolor="None")
blue_line = mlines.Line2D([], [], color='blue', markersize=5, label='Normal', linestyle=":", marker="8", markerfacecolor="None")

handles = [red_line,green_line, blue_line]
labels = [h.get_label() for h in handles] 

fig.legend(handles=handles, labels=labels, borderaxespad=5, title="Sheet", fancybox=True)

red_line_k = mlines.Line2D([], [], color='red',markersize=5, label='Group 1', linestyle=":", marker="s", markerfacecolor="None")
green_line_k = mlines.Line2D([], [], color='green', markersize=5, label='Group 2', linestyle=":", marker="s", markerfacecolor="None")
blue_line_k = mlines.Line2D([], [], color='blue', markersize=5, label='Group 3', linestyle=":", marker="s", markerfacecolor="None")

handles_k = [red_line_k,green_line_k, blue_line_k]
labels_k = [h.get_label() for h in handles_k] 

fig.legend(handles=handles_k, labels=labels_k, borderaxespad=12, title="Cluster k-mean", fancybox=True)

plt.show()
```
